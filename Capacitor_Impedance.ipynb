# Visualizing The Impedance equation for capacitors:




\begin{equation*}Imp = \frac{1}{2*pi*frequency*capacity}
\end{equation*}




<img src="./capacitor_impedance.png" width="600" />

#### These plots were generated with this Python code

````python
import matplotlib.pyplot as plt
from labellines import labelLine, labelLines
def capacitor_impedance():
    '''Generates a family of Impedance curves over a range of Capacitances and Frequencies'''
    pi=3.14159265
    colors = ['C' + str(n) for n in range(12)]
    capacities = [ pow(10,-n) for n in range(3,13)]
    labels =  ['1p', '10p', '100p', '1n', '10n', '100n', '1u', '10u', '100u', '1m']
    x = [ .1*k*(10**i) for i in range(0, 10) for k in range(1, 11, 1) ]
    for c in capacities:
        y= [  1/(2*pi*f*c)/1000 for f in x ]
        plt.xscale('log')
        plt.ylim(0, 2.5)
        line, = plt.plot(x, y, color=colors.pop(0), lw=2, label=labels.pop())
    plt.xlabel("Frequency")
    plt.ylabel("Impedance (K)")
    plt.title("Impedance Vs Capacity & Frequency")
    labelLines(plt.gca().get_lines(), align=False, color='k', fontsize=8)
    plt.show()
capacitor_impedance()
````
